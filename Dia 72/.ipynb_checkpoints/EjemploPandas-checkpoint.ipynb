```python
import pandas as pd

df = pd.read_csv("salaries_by_college_major.csv")

df.shape

```

```python
df.columns
```

```python
df.head()
```

```python
## Observo si alguno de los valores es distinto a un numero
df.isna()
```

```python
## Borro aquellos que no sean numeros

new_df = df.dropna()
```

```python
#What college major has the highest mid-career salary? How much do graduates with this major earn? (Mid-career is defined as having 10+ years of experience).

idx_max = new_df["Mid-Career Median Salary"].idxmax()
new_df.loc[idx_max]
```

```python
#Which college major has the lowest starting salary and how much do graduates earn after university?

idx_min = new_df["Starting Median Salary"].idxmin()
new_df.loc[idx_min]
```

```python
#Which college major has the lowest mid-career salary and how much can people expect to earn with this degree? 

idx_min = new_df["Mid-Career Median Salary"].idxmin()
new_df.loc[idx_min]
```

```python
spread_col = new_df["Mid-Career 90th Percentile Salary"] - new_df['Mid-Career 10th Percentile Salary']
new_df.insert(1,"Spread",spread_col)
```

```python
new_df.head()
```

```python
idx_min = new_df["Spread"].idxmin()
new_df.loc[idx_min]
```

```python
#Using the .sort_values() method, can you find the degrees with the highest potential? Find the top 5 degrees with the highest values in the 90th percentile. 
buf =new_df.sort_values(ascending=False, by="Mid-Career 90th Percentile Salary")
buf[["Undergraduate Major","Spread"]].head(5)
```

```python

#Also, find the degrees with the greatest spread in salaries. Which majors have the largest difference between high and low earners after graduation.
buf =new_df.sort_values(ascending=False, by="Spread")
buf[["Undergraduate Major","Spread"]].head(5)

```
